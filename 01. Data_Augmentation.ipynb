```python
import numpy as np
import os
import cv2
```

```python
# Read Images To Resize

for img in os.listdir("imgs"):
  image_path = os.path.join("imgs", img)
  myImg = cv2.imread(image_path)

  resized = cv2.resize(myImg, (128, 128), interpolation = cv2.INTER_LINEAR)

  cv2.imwrite(f"resized_imgs/{img}", resized)
```

```python
# Vertical Filp + Gaussian Noise + Color => Create new images

for img in os.listdir("resized_imgs"):
  image_path = os.path.join("resized_imgs", img)
  myImg = cv2.imread(image_path)

  flipped = cv2.flip(myImg, 1)
  noise_image = np.clip(flipped + np.random.normal(0, 15, flipped.shape), 0, 255).astype(np.uint8)
  changeColors = cv2.convertScaleAbs(noise_image, alpha = 1.2, beta = 30)
  resized = cv2.resize(changeColors, (128, 128), interpolation = cv2.INTER_LINEAR)

  cv2.imwrite(f"flip_noise_color_imgs/{img}", resized)
```

```python
# Horizontal Filp + Gaussian Blur => Create new images

for img in os.listdir("resized_imgs"):
  image_path = os.path.join("resized_imgs", img)
  myImg = cv2.imread(image_path)

  flipped = cv2.flip(myImg, 0)
  blurred = cv2.GaussianBlur(flipped, (5, 5), 0)
  resized = cv2.resize(blurred, (128, 128), interpolation = cv2.INTER_LINEAR)

  cv2.imwrite(f"flip_blur_imgs/{img}", resized)
```
